#**Question No 1.**
A weather station recorded the temperature (°C) of a city for 10 days:
 [30, 32, 31, 29, 35, 36, 34, 33, 31, 30]
Perform the following tasks:
*   a) Create a 1D NumPy array using the given data.
*    b) Display the temperature on the 5th day using indexing.
*   c) Display the last day temperature using negative indexing
*    d) Use slicing to display temperatures from day 3 to day 7.
*    e)Display temperatures greater than 33°C.







 .




```python
import numpy as np
```

```python
#a. Create a 1D NumPy array using the given data.
temperature = np.array([30,32,31,29,35,36,34,33,31,30])
print("Temperature Array:")
print(temperature)
```

```python
#b. Display the temperature on the 5th day using indexing.
print("\ntemperature on 5th Day:")
print(temperature[4])
```

```python
#c. Display the last day temperature using negative indexing
print("\nLast day Temprature:")
print(temperature[-1])
```

```python
#d. Use slicing to display temperatures from day 3 to day 7.
print("\nTemperatures from Day 3 to Day 7:")
print(temperature[2:7])
```

```python
#e. Display temperatures greater than 33°C.
print("\nTemperatures Greater than 33°C:")
print(temperature[temperature>33])
```

#**Question 2.**
Marks of 5 students in 4 subjects are given below:
Student  math  Science  English  Computer
S1       78    85       82       90
S2       88    79       91       86
S3       67    73       70       75
S4       92    88       84       89
S5       81    77       79       83
Perform the following tasks:

* Create a 2D NumPy array using the given data.
* Display marks of the 3rd student using indexing.
* Display marks of the first two subjects for all students using slicing.
* Display marks of the last student using negative indexing.
* Display marks of S2 and S3



.





```python
# a. Create a 2D NumPy array using the given data.
marks = np.array([[78,85,82,90],
                  [88,79,91,86],
                  [67,73,70,75],
                  [92,88,84,89],
                  [81,77,79,83]])
print("Student marks:")
print(marks)
```

```python
# b.Display marks of the 3rd student using indexing.
print("\nMarks of 3rd Student:")
print(marks[2])
```

```python
# c) Display marks of the first two subjects for all students
print("\nFirst Two Subjects of All Student:")
print(marks[:,0:2])
```

```python
# d) Display marks of the last student using negative indexing
print("\nMarks of Last Student:")
print(marks[-1])
```

```python
# e) Display marks of S2 and S3
print("\nMarks of S2 and S3:")
print(marks[1:3])
```

#**Question 3**
 A company recorded sales of 3 products for 4 months in 2 regions.

Product 1
Jan: [120,150]
Feb: [130,140]
Mar: [135,160]
Apr: [150,170]

Product 2
Jan: [200,210]
Feb: [195,205]
Mar: [220,230]
Apr: [210,225]

Product 3
Jan: [95,100]
Feb: [110,120]
Mar: [105,115]
Apr: [130,140]
Perform the following tasks:

a) Create a 3D NumPy array representing the sales data

b) Display the sales data of the 2nd product using indexing.

c) Display Region 1 sales for all products using slicing.

d) Display March sales for all products.


```python
#a) Create a 3D NumPy array representing the sales data
Sales = np.array([
#P1
[
    [120,150],
[130,140],
[135,160],
[150,170]],
 [
    #P2
[200,210],
[195,205],
[220,230],
[210,225]
    ],
   [ #P3
[95,100],
[110,120],
[105,115],
[130,140]
     ]
                   ])
print("Sales Data:")
print(Sales)
```

```python
# b) Display sales data of the 2nd product
print("\nSales Data of Product 2:")
print(Sales[1])
```

```python
# c) Display Region 1 sales for all products
print("\nRegion 1 Sales for All Product:")
print(Sales[:, :, 0])
```

```python
# d) Display March sales for all products
print("\nMarch Sales for All Products:")
print(Sales[:, 2, :])
```
