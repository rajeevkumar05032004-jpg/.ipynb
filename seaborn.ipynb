### **What is Seaborn?**

Seaborn is a high-level statistical data visualization library built on top of Matplotlib.

* Simplifies complex visualizations

* Works directly with Pandas DataFrames

* Provides attractive default styles

### **Installing & Importing Seaborn**

```python
import seaborn as sns
import pandas as pd
import numpy as np

```

```python
sns.get_dataset_names()
```

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
penguins["species"].value_counts()
```

```python
penguins["island"].value_counts()
```

```python
sns.scatterplot(data=penguins, x = "flipper_length_mm", y = "body_mass_g", hue = "sex" )
```

###**Seaborn Styling**

```python
sns.set_style("darkgrid")
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue = "island")
sns.despine(left = True)
```

### **Seaborn Distribution Plots**
A distribution plot shows how data values are spread across a range.

It helps understand:

Shape of data

* Shape of data
* Central tendency
* Spread
* Skewness


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data = np.random.rand(1000)
sns.set_style("darkgrid")
print(data)
```

```python
sns.histplot(data, kde=True, bins=30, color='skyblue')
plt.title("Seaborn Distribution Plot")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
